## 2. The Aiyagari (1994) Model

### Pregunta 2.1

Ahora resolveremos una instancia del problema de Aiyagari. En este modelo el consumo $c_t$ es el numerario.

El problema del hogar se ve como:

$$Max{ \mathbb{E}\sum_{t=0}^{\infty} \beta^{t}\log{c_t}} $$

$$s.t.: c_t+k_{t+1} \leq (1+r_t-\delta)k_t+w_texp(z_t)\bar{h}$$

$$ k_{t+1}\geq \underline{b}$$

$$ z_{t+1}= \rho_z z_t+\epsilon_{z,t+1}$$

La deuda se puede reescribir como:

$$k_{t+1}\geq -\phi$$ 

donde  $\phi=min\lbrace\underline{b}, \frac{wl_{min}}{r}\rbrace$

El capital total en un periodo de tiempo puede ser propio y prestado:

$$\hat{k}_{t+1}= k_{t+1}+\phi$$

El total de activos del hogar en un periodo $t$ es:

$$A_t=w_texp(z_t)\bar{h}+(1-r_t-\delta) \hat{k}_t - (r_t-\delta)\phi$$

si despejamos $k_t$ de la definicion de $\hat{k}_t$ podemos escribir la restriccion como:

$$ c_t+\hat{k}_{t+1}-\phi = (1+r_t-\delta)(\hat{k}_t-\phi)+w_texp(z_t)\bar{h}$$

Ordenando nos queda:

$$ c_t+\hat{k}_{t+1}= (1+r_t-\delta)\hat{k}_t+w_texp(z_t)\bar{h}-(r_t-\delta)\phi$$

El lado derecho de la igualdad es $A_t$, por lo tanto:

$$c_t+\hat{k}_{t+1}= A_t $$

o lo que es igual:

$$c_t= A_t - \hat{k}_{t+1}$$

Dado esto, podemos escribir el problema dinamico con la siguiente ecuacion de Bellman:

$$V(A_t,\underline{b},w,r)(A_{t+1})= Max{\lbrace U(A_t-\hat{k}_{t+1})+\beta \int{V(A_{t+1}, \underline{b},w,r)\partial F}(z_{t+1})\rbrace}$$

$$s.t.: A_t=w_texp(z_t)\bar{h}+(1-r_t-\delta) \hat{k}_t - (r_t-\delta)\phi $$

$$z_{t+1}= \rho_z z_t+\epsilon_{z,t+1} $$

Cambiando $t+1=\prime$ tenemos

$$V(A,\underline{b},w,r)(A^\prime)= Max{\lbrace U(A-\hat{k}^\prime)+\beta \int{V\partial F}(z^\prime)\rbrace}$$

$$s.t.: A=w_texp(z)\bar{h}+(1-r-\delta) \hat{k} - (r-\delta)\phi $$

$$z^\prime= \rho_z z+\epsilon_{z^\prime} $$

Vemos que la variable de control es el capital $\hat{k}_{t+1}= k_{t+1}+\phi $ y la variable de estado son los activos: $A_t=w_texp(z_t)\bar{h}+(1-r_t-\delta) \hat{k}_t - (r-\delta)\phi $ 

### Pregunta 2.2

Las CPO las podemos obtener del problema original:

$$\mathcal{L}=\mathbb{E}\sum_{t=0}^{\infty} \beta^{t}\log{c_t}+\lambda_t[(1+r_t-\delta)k_t+w_texp(z_t)\bar{h}-c_t-k_{t+1}]$$

Las CPOS son:

$$\frac{\partial\mathcal{L}}{\partial c_t}= \frac{\beta^t}{c_t}-\lambda_t=0$$

donde obtenemos que:

$$\frac{\beta^t}{c_t}=\lambda_t$$ y $$\frac{\beta^{t+1}}{c_{t+1}}=\lambda_t$$

$$\frac{\partial\mathcal{L}}{\partial k_t}= \lambda_t (1+r_t-\delta)-\lambda_{t+1}=0$$

sustituyendo $\lambda_t$ y $\lambda_{t+1}$ tenemos:

$$\frac{c_t}{c_{t+1}}=\frac{1}{\beta}(1+r_t-\delta)$$ que es nuestra ecuacion de euler.

Las condiciones de primer orden nos dicen como se comporta la familia en el margen. Basicamente, permite establecer una policy hacia adelante actuando en funcion de cual es el beneficio marginal mas grande que pueden obtener actuando optimamente hacia adelante. Tambien muestra la oferta de capital de los hogares.

### Pregunta 2.3

El problema de la firma se ve como:

$$ max {\Pi= K_t^\alpha L_t^{1-\alpha}-r_t K_t-w_t L_t} $$

las cpos de la firma son:

$$\frac{\partial\mathcal{\Pi}}{\partial K_t}= \alpha K_t^{\alpha-1} L_t^{\alpha}-r_t= 0$$

Despejando nos queda:

$$r_t=\alpha (\frac{L_t}{K_t})^{1-\alpha}$$

$$\frac{\partial\mathcal{\Pi}}{\partial L_t}= (1-\alpha)K_t^\alpha L_t^{-\alpha}-w_t= 0$$

despejando y reordenando:

$$ w_t= (1-\alpha)(\frac{L_t}{K_t})^{\alpha}$$

$$ w_t= (1-\alpha)(\frac{\alpha}{r_t})^{\frac{\alpha}{\alpha-1}}$$

Las condiciones de primer orden nos dicen como se comporta la firma en el margen. Basicamente, permite establecer una policy hacia adelante actuando en funcion de cual es el beneficio marginal mas grande que pueden obtener actuando optimamente hacia adelante. Tambien muestra cual es la demanda de capital y trabajo.

```python
df=pd.read_stata("JPSAERdata.dta") # Open database Household Expenditure and the Income Tax Rebates of 2001 
                                   #(Johnson, Parker, and Souleles, 2001)

df=df.join(pd.get_dummies(df['yymm'])) #Create dummies for each period of time

#Create cuts for income and liquidity. We use the 30 and 70 percentiles to set the limits.

low_liquid= (df[["liqassii"]][(df["dcn"]!=None) & 
                              (df["sample"]>0) & 
                              (df["taxreb"]!=None) & 
                              (df["taxreb"]>0) & 
                              (df["ltaxreb"]!=None)].quantile([0.3,0.7])).iloc[0,0]
high_liquid=(df[["liqassii"]][(df["dcn"]!=None) & 
                              (df["sample"]>0) & (df["taxreb"]!=None) & 
                              (df["taxreb"]>0) & (df["ltaxreb"]!=None)].quantile([0.3,0.7])).iloc[1,0]

df["Low Liquid"]=df["liqassii"][df["liqassii"].notnull()].apply(lambda x: 1 if x<=low_liquid else 0)
df["High Liquid"]= df["liqassii"][df["liqassii"].notnull()].apply(lambda x: 1 if x>high_liquid else 0)



low_income= (df[["income"]][(df["dcn"]!=None) &
                            (df["sample"]==1) &
                            (df["taxreb"]!=None) &
                            (df["taxreb"]>0) &
                            (df["ltaxreb"]!=None)].quantile([0.3,0.7])).iloc[0,0]
high_income=(df[["income"]][(df["dcn"]!=None) &
                            (df["sample"]==1) &
                            (df["taxreb"]!=None) &
                            (df["taxreb"]>0) &
                            (df["ltaxreb"]!=None)].quantile([0.3,0.7])).iloc[1,0]

df["Low Income"]=df["income"][df["income"].notnull()].apply(lambda x: 1 if x<=low_income else 0)
df["High Income"]= df["income"][df["income"].notnull()].apply(lambda x: 1 if x>high_income else 0)

#Define Poor hand-to-mouth and Wealthy hand-to-mouth

df["Poor_hand_to_mouth"]= df["Low Income"]*df["Low Liquid"]
df["Wealthy_hand_to_mouth"]= df["High Income"]*df["Low Liquid"]

#Create de interactions variables

df["Poor_hand_to_mouth_tax_reb"]=df["taxreb"]*df["Poor_hand_to_mouth"]
df["Poor_hand_to_mouth_tax_reb_dummy"]=df["itaxreb"]*df["Poor_hand_to_mouth"]
df["Poor_hand_to_mouth_lag_tax_reb"]=df["ltaxreb"]*df["Poor_hand_to_mouth"]
df["Poor_hand_to_mouth_lag_tax_reb_dummy"]=df["iltaxreb"]*df["Poor_hand_to_mouth"]
df["Wealthy_hand_to_mouth_tax_reb"]=df["taxreb"]*df["Wealthy_hand_to_mouth"]
df["Wealthy_hand_to_mouth_tax_reb_dummy"]=df["itaxreb"]*df["Wealthy_hand_to_mouth"]
df["Wealthy_hand_to_mouth_lag_tax_reb"]=df["ltaxreb"]*df["Wealthy_hand_to_mouth"]
df["Wealthy_hand_to_mouth_lag_tax_reb_dummy"]=df["iltaxreb"]*df["Wealthy_hand_to_mouth"]

#Define the total number positives rebates

household_positive_rebate=df["taxreb"][(df["taxreb"]>0) & (df["sample"]>0)].count()

#Replicating table 1 panel b

df_table1b=pd.DataFrame(pd.concat(
    [df[["taxreb"]][(df["taxreb"]>0) & (df["sample"]>0) & (df["taxreb"]<300)].count(),
     df[["taxreb"]][(df["taxreb"]==300) & (df["sample"]>0)].count(),
     df[["taxreb"]][(df["taxreb"]>300) & (df["sample"]>0) & (df["taxreb"]<600)].count(),
     df[["taxreb"]][(df["taxreb"]==600) & (df["sample"]>0)].count(),
     df[["taxreb"]][(df["taxreb"]>600) & (df["sample"]>0)].count()]))

df_table1b["% of Positives Rebates"]= df_table1b[0]/household_positive_rebate*100

df_table1b.rename(columns={0:"Number of Observations"}, inplace=True)

df_table1b.set_axis(["0 < Rebate < 300","Rebate = 300","300 < Rebate < 600","Rebate = 600","Rebate > 600"], 
                    axis=0, inplace=True)

df_table1b.index.rename("Rebate Value", inplace=True)

#Adding positive rebates for Poor hand-to-mouth and Wealthy hand-to-mouth

Poor_hand_to_mouth=(pd.concat(
    [df[["Poor_hand_to_mouth_tax_reb"]][(df["Poor_hand_to_mouth_tax_reb"]>0) & 
                                        (df["sample"]>0) & 
                                        (df["Poor_hand_to_mouth_tax_reb"]<300)].count(),
     df[["Poor_hand_to_mouth_tax_reb"]][(df["Poor_hand_to_mouth_tax_reb"]==300) & 
                                        (df["sample"]>0)].count(),
     df[["Poor_hand_to_mouth_tax_reb"]][(df["Poor_hand_to_mouth_tax_reb"]>300) & 
                    (df["sample"]>0) & 
                    (df["Poor_hand_to_mouth_tax_reb"]<600)].count(),
     df[["Poor_hand_to_mouth_tax_reb"]][(df["Poor_hand_to_mouth_tax_reb"]==600) & 
                    (df["sample"]>0)].count(),
     df[["Poor_hand_to_mouth_tax_reb"]][(df["Poor_hand_to_mouth_tax_reb"]>600) & 
                    (df["sample"]>0)].count()]))

df_table1b["Number of Poor hand-to-mouth Positives Rebates"]=Poor_hand_to_mouth.values

df_table1b["% of Poor hand-to-mouth Positives Rebates"]= (df_table1b["Number of Poor hand-to-mouth Positives Rebates"]/
                                                          household_positive_rebate*100)

Wealthy_hand_to_mouth=(pd.concat(
    [df[["Wealthy_hand_to_mouth_tax_reb"]][(df["Wealthy_hand_to_mouth_tax_reb"]>0) & 
                                        (df["sample"]>0) & 
                                        (df["Wealthy_hand_to_mouth_tax_reb"]<300)].count(),
     df[["Wealthy_hand_to_mouth_tax_reb"]][(df["Wealthy_hand_to_mouth_tax_reb"]==300) & 
                                        (df["sample"]>0)].count(),
     df[["Wealthy_hand_to_mouth_tax_reb"]][(df["Wealthy_hand_to_mouth_tax_reb"]>300) & 
                    (df["sample"]>0) & 
                    (df["Wealthy_hand_to_mouth_tax_reb"]<600)].count(),
     df[["Wealthy_hand_to_mouth_tax_reb"]][(df["Wealthy_hand_to_mouth_tax_reb"]==600) & 
                    (df["sample"]>0)].count(),
     df[["Wealthy_hand_to_mouth_tax_reb"]][(df["Wealthy_hand_to_mouth_tax_reb"]>600) & 
                    (df["sample"]>0)].count()]))

df_table1b["Number of Wealthy hand-to-mouth Positives Rebates"]=Wealthy_hand_to_mouth.values

df_table1b["% of Wealthy hand-to-mouth Positives Rebates"]= (
    df_table1b["Number of Wealthy hand-to-mouth Positives Rebates"]/household_positive_rebate*100)

df_table1b.loc["Total"]= df_table1b.sum().values

print(df_table1b.apply(lambda x: round(x,1)).to_markdown())
```

```python
df=pd.read_stata("JPSAERdata.dta") #Resetting again the original data frame

df=df.join(pd.get_dummies(df['yymm'])) #Create dummies for each period of time

#Creating functin able to open databses and create a DataFrame with the columns of interest

def file_reader(file_list, column_input):
    df = pd.DataFrame()

    columns_to_read = ["NEWID"]
    columns_to_read.extend(column_input)
    for file in file_list:
        try:
            temp = pd.read_csv(file)
            temp = temp[columns_to_read]
        except FileNotFoundError:
            print(f"{file} not found")
            break
        except KeyError:
            print(f"{column_input} not found in {file}")
            break
        else:
            temp["newid"] = temp.NEWID.apply(lambda x: int(str(x)[:-1]))
            temp["intview"] = temp.NEWID.apply(lambda x: int(str(x)[-1]))
            df = pd.concat([df, temp], axis=0)
            
    df= df.drop_duplicates(subset=["newid", "intview"], keep="last")

    return df

#Creating function that merges all the dataframes to the main dataframe

def df_joiner(left, dfs_to_join):

    for df in dfs_to_join:
        left = left.merge(df, on=["newid", "intview"], how = "left")

    return left

#Files to open

files=["mor01.csv", "mor01.csv"]

df_mor= file_reader(files, ["ORGMRTX"])

files= ["opb01.csv", "opb02.csv"]

df_opb= file_reader(files, ["PROPVALX"])

files= ["fmli011x.csv","fmli012.csv","fmli013.csv","fmli014.csv","fmli021.csv", "fmli022.csv",
       "fmli023.csv", "fmli024.csv", "fmli031.csv"]

df_fmli= file_reader(files, ["CKBKACTX", "SAVACCTX", "USBNDX", "SECESTX"])

dataframes= [df_mor, df_opb, df_fmli]
df= df_joiner(df, dataframes)

#replacing missing values with 0

df= df.replace(".", 0)

#Creating variable Total Wealth.

df["Total Wealth"]= (df["CKBKACTX"].astype(float)+ df["SAVACCTX"].astype(float)+
                     df["USBNDX"].astype(float)+ df["SECESTX"].astype(float)+ 
                     df["PROPVALX"].astype(float) - df["ORGMRTX"].astype(float))
```

```python
class Household:
    """
    This class takes the parameters that define a household asset accumulation
    problem and computes the corresponding reward and transition matrices R
    and Q required to generate an instance of DiscreteDP, and thereby solve
    for the optimal policy.

    Comments on indexing: We need to enumerate the state space S as a sequence
    S = {0, ..., n}.  To this end, (a_i, z_i) index pairs are mapped to s_i
    indices according to the rule

        s_i = a_i * z_size + z_i

    To invert this map, use

        a_i = s_i // z_size  (integer division)
        z_i = s_i % z_size

    """


    def __init__(self,
                r=0.0225,                      # Interest rate
                w=1.0,                         # Wages
                =1/1.039,                     # Discount factor
                a_min=1e-10,
                a_max=18,
                a_size=200,
                =0.74,                        #AR1 persistence process 
                =0.78,                        #AR1 Variance process
                =0.0125):                     #Depreciation


        # Store values, set up grids over a and z
        self.r, self.w, self., self., self., self. = r, w, , , , 
        self.a_min, self.a_max, self.a_size = a_min, a_max, a_size
        
        
        self. = qe.rouwenhorst(rho=, ybar=0, sigma= , n=2).P #Markov Transition matrix
        self.z_vals = np.exp(qe.rouwenhorst(rho=, ybar=0, sigma= , n=2).state_values) #Markov State Values
        self.z_size = len(self.z_vals)
        self.h= 0.33
        # Hours worked
        

        self.a_vals = np.linspace(a_min, a_max, a_size)
        self.n = a_size * self.z_size

        # Build the array Q
        self.Q = np.zeros((self.n, a_size, self.n))
        self.build_Q()

        # Build the array R
        self.R = np.empty((self.n, a_size))
        self.build_R()

    def set_prices(self, r, w):
        """
        Use this method to reset prices. Calling the method will trigger a
        re-build of R.
        """
        self.r, self.w = r, w
        self.build_R()

    def build_Q(self):
        populate_Q(self.Q, self.a_size, self.z_size, self.)

    def build_R(self):
        self.R.fill(-np.inf)
        populate_R(self.R,
                   self.a_size,
                   self.z_size,
                   self.a_vals,
                   self.z_vals,
                   self.r,
                   self.w,
                   self.,
                   self.h)


# Do the hard work using JIT-ed functions

@jit(nopython=True)
def populate_R(R, a_size, z_size, a_vals, z_vals, r, w, , h):
    n = a_size * z_size
    for s_i in range(n):
        a_i = s_i // z_size
        z_i = s_i % z_size
        a = a_vals[a_i]
        z = z_vals[z_i]
        for new_a_i in range(a_size):
            a_new = a_vals[new_a_i]
            c = w * z * h + (1 + r -  ) * a - a_new
            if c > 0:
                R[s_i, new_a_i] = np.log(c)  # Utility

@jit(nopython=True)
def populate_Q(Q, a_size, z_size, ):
    n = a_size * z_size
    for s_i in range(n):
        z_i = s_i % z_size
        for a_i in range(a_size):
            for next_z_i in range(z_size):
                Q[s_i, a_i, a_i*z_size + next_z_i] = [z_i, next_z_i]


@jit(nopython=True)
def asset_marginal(s_probs, a_size, z_size):
    a_probs = np.zeros(a_size)
    for a_i in range(a_size):
        for z_i in range(z_size):
            a_probs[a_i] += s_probs[a_i*z_size + z_i]
    return a_probs
```
